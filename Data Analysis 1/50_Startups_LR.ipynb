# creating Model

```python
# %pip install scikit-learn
data.columns
```

```python
#  converting State into float
from sklearn.preprocessing import LabelEncoder as LE
le = LE()
data["State_float"] = le.fit_transform(data.State)
data.State_float 
```

```python
data.head(10)
```

```python

x = data[["R&D Spend","Administration","Marketing Spend","State_float"]]
# y = data["Profit"]
y = data.Profit

# x = input data   &   y = output data
```

```python
x.head()
```

```python
y.head()
```

```python
# training and testing creation
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y, test_size=0.25,random_state=42) 
#  test-size = how percent data you wanna test 1 = 100% , 0.25 = 25% --> default
```

```python
len(x_train) # 37
x_train.dtypes
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
# creating model 
# learn bg theory
from sklearn.linear_model import LinearRegression as LR
LR = LR() # object creation

LR.fit(x_train,y_train)  # training of linear regression


```

```python
y_predict = LR.predict(x_test)
data.head()
```

```python
y_predict
```

```python
y_test
```

```python
x1 =15000
x2 =10000
x3 =50000
x4 =1

LR.predict([[x1,x2,x3,x4]])
```

```python
m = LR.coef_
m[0]*x1 +  m[1]*x2 + m[2]*x3 + m[3]*x4 + LR.intercept_
m
```

```python
# Take user input for x1, x2, x3, x4
RW_spends = float(input("Enter R&D Spend: "))
Admin_spends = float(input("Enter Administration Spend: "))
Marketing_spends = float(input("Enter Marketing Spend: "))
State = float(input("Enter State (as float, e.g., 0, 1, or 2): "))

# Predict profit using the trained model
LR.predict([[RW_spends, Admin_spends, Marketing_spends, State]])


Profit = m[0]*RW_spends +  m[1]*Admin_spends + m[2]*Marketing_spends + m[3]*State + LR.intercept_
Profit


```

```python
#  model evaluation
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
r2score = r2_score(y_test,y_predict)

r2score
```

```python
x_test.shape
```

```python
'''
#  calc of r2score
# Adjusted R = 1 - ( (1 - R) * (n - 1) / (n - p - 1) )

R = R-squared value
n = number of samples (rows in your test set) of x_test
p = number of predictors/features

# adjusted_r2 = 1 - (( 1 - r2score ) * ( 13 - 1 ) * ( 13 - 4 - 1) )
'''
adjusted_r2 = 1 - (( 1 - r2score ) * ( x_test.shape[0] - 1 ) / ( x_test.shape[0] - x_test.shape[1] - 1) )
```

```python
adjusted_r2
```
